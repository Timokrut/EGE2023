```python
def count_div(n): #подсчет делителей
    k = 0
    for  j in range(1, int(n**0.5) + 1):
        if n % j == 0:
            k += 1
            if j != n // j:
                k += 1
    return k
def find_div(n): #найти делители
    a = []
    for j in range(1, int(n**0.5) + 1):
        if n % j == 0:
            a.append(j)
            if j != n // j:
                a.append(n//j)
    a.sort()
    return a   
def is_prime(n): #простое или нет
    if n <= 1:
        return False
    for j in range(2, int(n**0.5) + 1):
        if n % j == 0:
            return False
    return True
k = 0
ans = 0
for x in range(500001, 505000):
    y = find_div(x)[-2]
    k += 1
    if is_prime(y) == 0:
        ans += 1
        print(x, find_div(x)[-1])
        if ans == 7:
            break
```
